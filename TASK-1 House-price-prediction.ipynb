```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file
file_path = '/kaggle/input/housedata/output.csv'  # Adjust the path to your actual file location
data = pd.read_csv(file_path)

# Displaying basic information about the dataset
print("Dataset information:")
print(data.info())

# Visualizing data (you can customize this based on your specific dataset)
# Example: Plotting a histogram for a numerical column
plt.figure(figsize=(10, 6))
plt.hist(data['date'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of a Numerical Column')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Example: Scatter plot for two numerical columns
plt.figure(figsize=(10, 6))
plt.scatter(data['price'], data['bedrooms'], alpha=0.5, color='green')
plt.title('Scatter Plot of Two Numerical Columns')
plt.xlabel('house price on date')
plt.ylabel('house price based on bedromms')
plt.show()


```

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reading the CSV file
file_path = '/kaggle/input/housedata/data.csv'  # Adjust the path to your actual file location
data = pd.read_csv(file_path)

# Displaying basic information about the dataset
print("Dataset information:")
print(data.info())

# Scatter plot for two numerical columns using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='date', y='price', data=data, alpha=0.5, color='green')
plt.title('Scatter Plot of Two Numerical Columns (Seaborn)')
plt.xlabel('house prediction various on date')
plt.ylabel('house prediction various on price')
plt.show()
```

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Reading the CSV file using NumPy
file_path = '/kaggle/input/housedata/output.csv'  # Adjust the path to your actual file location
data = np.genfromtxt(file_path, delimiter=',')

# Displaying basic information about the dataset
print("Dataset information:")
print(data[:5])  # Displaying the first 5 rows as an example

# Scatter plot for two numerical columns using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(data[:, 0], data[:, 1], alpha=0.5, color='red')
plt.title('Scatter Plot of Two Numerical Columns (NumPy & Matplotlib)')
plt.xlabel('price prediction')
plt.ylabel('house neighborhood')
plt.grid(True)
plt.show()
```
